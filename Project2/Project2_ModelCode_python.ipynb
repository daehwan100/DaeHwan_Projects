# 데이터 진행 과정

```python
# os지정 및 데이터 확인
import os
for dirname, _, filenames in os.walk('C:/input/new_dataset'): #파일 경로 지정 필요
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
# image_dir에 지정한 위치 지정하기
image_dir = Path('C:/input/new_dataset/new_aug_Braille Dataset')

# dir_list에 지정한 위치 파일안에 jpg 주소들 저장하기
dir_list = list(image_dir.glob('*.jpg'))

# image_count에 파일 개수들 저장하기
image_count = len(dir_list)
image_count

# jpg주소를 str()에 넣어 images배열에 저장한다   
# imread함수는 이미지를 255X255XRGB(3)형태로 불러오는 것
images = []
for dir in dir_list:
    Add = cv2.imread(str(dir))
    images.append(Add)
    
# 각 파일이름 제일 앞 글자들을 출력값으로 지정해준다(name_list
name_list = []
for i in dir_list:
    name_list.append(os.path.basename(i)[0]) # 첫 알파벳만 가져온다. append는 list에 추가하는 함수

# imgaes빠른 연산과 메모리 효율을 위해 리스트를 넘파이 배열로 바꾼다
images_list = np.array(images)
# 파일 앞 글자 저장 리스트들을 transpose하여 행과 열을 바꾼다
name_list = np.array(name_list).T
# shape확인하기
images_list.shape
name_list.shape

# a = 0, b = 1, c = 2 ... z = 25 변환하기
le = LabelEncoder()
name_list = le.fit_transform(name_list) 

#  정규화
images_list = images_list / 255.0 

# 정규화 잘 됐는지 확인하기
print(images_list[0])


from sklearn.model_selection import train_test_split # 모듈 사용
# 전체데이터를 훈련데이터 85%, 시험데이터15%로 랜덤으로 나누어 저장
X_train, X_test, y_train, y_test = train_test_split(images_list, name_list, test_size=0.15, random_state=42)

# 위에서 나눠진 훈련데이터를 훈련데이터85%, 검증데이터15%로 랜덤으로 나누어 저장
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.15, random_state=42)
```

## 모델 구성

```python
# 모델 구성
model = keras.Sequential([
    keras.layers.Conv2D(filters=64, kernel_size=(5, 5), padding='same', activation='relu'), 
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.BatchNormalization(),

    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Dropout(0.2),
    keras.layers.BatchNormalization(),
    
    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Dropout(0.2),   
    keras.layers.BatchNormalization(),

    keras.layers.Flatten(),
    
    keras.layers.Dense(units=576, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.BatchNormalization(),

    keras.layers.Dense(units=288, activation="relu"),

    keras.layers.Dense(units=26, activation="softmax") 
])


# Adam 방법
model.compile(optimizer="Adam", loss="SparseCategoricalCrossentropy", metrics=["sparse_categorical_accuracy"])
"""
model.compile 변경수치 설명
RMS,SGD,Adam중에서는 SGD는 확실히 결과값이 좋지 않았고 RMS는 Adam에 비해 소요시간이 10초 늦었다. 따라서 최종적으로 Adam을 채택했다.
RMSprop 방법
#model.compile(optimizer="RMSprop", loss="SparseCategoricalCrossentropy", metrics=["sparse_categorical_accuracy"])
SGD 방법
#model.compile(optimizer="SGD", loss="SparseCategoricalCrossentropy", metrics=["sparse_categorical_accuracy"])
"""

# 모델학습
history = model.fit(x=X_train, y=y_train, epochs = 10, batch_size = 128, validation_data=(X_validation, y_validation))

# 학습진행과정
print(history.params)

# 모델요약
model.summary()

# 훈련데이터 정확도값들 저장
acc = history.history['sparse_categorical_accuracy']

# 검증데이터 정확도값들 저장
val_acc = history.history['val_sparse_categorical_accuracy']

# 훈련데이터 손실값들 저장
loss = history.history['loss']

# 검증데이터 손실값들 저장
val_loss = history.history['val_loss']

# 에폭수 범위저장
epochs = range(len(acc))
```

## 그래프 확인

```python
# 훈련데이터 검증데이터 정확도 그래프 나타내기
plt.plot(epochs, acc, 'b', label='Training accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()

# 훈련데이터 검증데이터 손실함수 그래프 나타내기
plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

## 정확도 확인

```python
# 훈려데이터 정확도
acc_train = model.evaluate(X_train, y_train, verbose=0)
print("Training Accuracy: ", round(acc_train[1]*100, 5), "%")

# 검증데이터 정확도
acc_val = model.evaluate(X_validation, y_validation, verbose=0)
print("Validation Accuracy: ", round(acc_val[1]*100, 5), "%")

# 시험데이터 정확도
acc_test = model.evaluate(X_test, y_test, verbose=0)
print("Test Accuracy: ", round(acc_test[1]*100, 5), "%")
```
